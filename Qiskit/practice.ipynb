```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import random_statevector, partial_trace, state_fidelity
```

```python
circ = QuantumCircuit(1, 1)

circ.draw(output='mpl')
```

```python
circ2 = QuantumCircuit(3, 3)
circ2.measure(range(3), range(3))
circ2.draw(output='mpl')
```

```python
backend_sim = Aer.get_backend('qasm_simulator')
job_sim = backend_sim.run(circ2, shots = 512)
result_sim = job_sim.result()
counts = result_sim.get_counts(circ2)
```

```python
type(counts)
```

```python
print(counts)
plot_histogram(counts)
```

```python
circ3 = QuantumCircuit(1, 1)
circ3.h(0)
circ3.measure(range(1), range(1))

job_2_sim = backend_sim.run(circ3, shots=1024)
counts = job_2_sim.result().get_counts()

print(counts)
plot_histogram(counts)
```
